<h1>Business Understanding 

To increase its loan business,the Universal Bank wants to predict the customers which have the higher probablity to purchase the loan depending on the various features.

<h1>Understanding the Dataset

****
This dataset contains the data of the customers of the Universal Bank.<br>
It contains the various features:<br>
1. ID-The Id of the customers which uniquely identifies each customer.<br>
2. Age-Age of the customer.<br>
3. Experience-Work Experience of customer.<br>
4. Income-Yearly income of the customer(every 10000).<br>
5. ZIP Code-Zip code of the address of the customer.<br>
6. Family-Number of family members.<br>
7. CCAvg-Average of Credit Card Yearly Expenses.<br>
8. Education-Education of the customer.<br>
       where 1-Level 1
             2-Level 2
             3-Level 3
       
9. Mortgage-Value of house mortgage.<br>
10. Securities-Does the customer have a securities account with the bank? (1=Yes, 0=No)<br>
11. CDAccount-Does the customer have a certificate of deposit with the bank? (1=Yes, 0=No)<br>
12. Online-Does the customer use Net banking facilities (1=Yes, 0=No)<br>
13. CreditCard-Does the customer use a credit card? (1=Yes, 0=No)<br>

<h1>Data Preprocessing

```python
X=np.asarray(df[['Age','Experience','Income','Family','CCAvg','Mortgage','Education']])
y=np.asarray(df['Personal Loan'])
print(X[0:10])
print(y[0:10])
```

<h2>Train/Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.24,random_state=4)
print("Train set:",X_train.shape,y_train.shape)
print("Test set:",X_test.shape,y_test.shape)
```

We can clearly see our training and testing data.

<h2>Support Vector System

```python
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(X_train, y_train) 
```

```python
#Predicting the data 
yhat=clf.predict(X_test)
```

Using Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,classification_report
con=confusion_matrix(y_test, yhat)
print(con)
```

```python
from mlxtend.plotting import plot_confusion_matrix as pc
plo=pc(conf_mat=con,cmap='winter_r',figsize = (5, 5))
```

Hence this confusion matrix plot shows the correct and incorrect predictions that our model has made.

```python
print(classification_report(y_test, yhat))
```

Out model has a 0.92 precision in predicting the customers not taking a loan and 0.62 precision in predicting customers taking a loan.<br>
There are many other factors calculation like f1-score.

<h1>Logistic Regression

```python
#Normalize the dataset
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.285,random_state=4)
print("Train set:",X_train.shape,y_train.shape)
print("Test set:",X_test.shape,y_test.shape)
```

Modelling

```python
from sklearn.linear_model import LogisticRegression
Lr=LogisticRegression(C=0.01,solver='liblinear').fit(X_train,y_train)
Lr
```

Predicting the data

```python
yhat=Lr.predict(X_test)
print(yhat)
```

Using the confusion matrix

```python
con_lr=confusion_matrix(y_test, yhat)
print(con_lr)
```

```python
from mlxtend.plotting import plot_confusion_matrix as pc
plo=pc(conf_mat=con_lr,cmap='winter_r',figsize = (5, 5))
```

This plot shows us the various prediction that our model has correctly and incorrectly predicted. 

```python
print(classification_report(y_test, yhat))
```

Using Logistic Regression,our model has a 0.94 precision in predicting the customers having not taking a loan and 0.89 precision for the customers taking loans.
